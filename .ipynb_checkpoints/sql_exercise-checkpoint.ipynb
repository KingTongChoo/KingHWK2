#### 1) Open the database:

```python
import sqlite3
import pandas as pd
db = sqlite3.connect('flights.db')
pd.read_sql("""SELECT * FROM airports""", con = db)
#I manually moved the flights.db from the assets value to the practice folder bcauase assets/flights.db returned an error
```

### Airports

#### 2) How many airports are there altogether?

```python
pd.read_sql("""
    SELECT
        COUNT(name)
    FROM
        airports
    """, con=db)
```

#### 3) How many airports are there in Morocco?

```python
pd.read_sql("""
    SELECT
        count(name),
        country
    FROM 
        airports
    WHERE 
       country = 'Morocco'
        """, con = db)
```

#### 4) Select the name and coordinates of the airport that has the code `HOT`

```python
#return the lat/longit HOT
pd.read_sql("""
    SELECT
        name,
        latitude,
        longitude
    FROM
        airports
    WHERE 
        code == 'HOT'
    """, con = db)
```

#### 5) Which country has the most airports?

```python
pd.read_sql("""
    SELECT
        country,
        count(name)
    FROM
        airports
    group by 
        name
    ORDER BY 
        count(name) desc
        """,con = db)
```

#### 6) What are the highest and lowest altitudes at which there are airports?

```python
pd.read_sql("""
    SELECT
        MAX(altitude),
        MIN(altitude)
    FROM 
        airports
    """, con = db)
```

## Airlines

#### 7) Select the last 5 airlines alphabetically

```python
pd.read_sql("""
    SELECT
        name
    FROM
        airlines
    ORDER BY 
        name      
""", con = db)
```

#### 8) Which country has the most *inactive* airlines?

```python
pd.read_sql("""
    select
        COUNTRY,
        count(active)
    from 
        airlines
    where 
        active = 'N'
    group by 
        country
    order by 
        count(active) desc
    """, con = db)

#how does one add an extra layer of condiiton?
```

#### 9) Which countries have *exactly* 9 active airlines?

Hint: look at the [HAVING clause](https://www.w3schools.com/sql/sql_having.asp)

```python
pd.read_sql("""
    SELECT
        country,
        count(active)
    FROM 
        airlines
    WHERE 
        active = 'Y'
    group by 
        country
    having 
        count(active) = 9
    """, con=db)
```

#### 10) Which airlines contain the word 'green' somewhere in their name?

Hint: look at the [LIKE](https://www.w3schools.com/sql/sql_like.asp) keyword to see how to search in text columns

```python
pd.read_sql("""
    SELECT
        name
    FROM
        airlines
    WHERE name like '%green%'
    """, con = db)
```

## Routes

#### 11) Select the first 10 routes

```python
pd.read_sql("""
    SELECT
        *
    FROM
        routes
    LIMIT 
        10
    """, con=db)
```

#### 12) Select the first 10 routes as well as the *country of the origin airport*

```python
pd.read_sql("""
    SELECT
        source as origin,
        dest as destination
    FROM 
        routes
    LIMIT
        10
        """, con = db)
```

```python
pd.read_sql("""
    SELECT
        source as origin,
        dest as destination
    FROM 
        routes
    LIMIT
        10
        """, con = db)
```

#### 14) How many routes are there *out* of Japan?

```python
pd.read_sql("""
    SELECT
        count(source)
    FROM
        routes
    WHERE
        source = 'JPA'
        """, con=db)
```

#### 15) How many routes are there *into* Brazil?

```python
pd.read_sql("""
    SELECT
        count(dest)
    FROM
        routes
    WHERE
        dest = 'BRA'""", con=db)
```
